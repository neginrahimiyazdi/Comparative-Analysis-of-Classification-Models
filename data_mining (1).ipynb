Label Encoding vs One_hot Encoding


*One* Hot Encoding:

در این روش، برای هر دسته از داده‌های رشته‌ای یک ویژگی جدید ایجاد می‌کنیم و برای هر نمونه، ویژگی مربوط به دسته‌ی آن نمونه را 1 و بقیه ویژگی‌ها را 0 قرار می‌دهیم. این روش مناسب است وقتی ترتیب دسته‌ها اهمیتی ندارد و می‌خواهیم مدل ماشین‌لرنینگ به این ترتیب حساس نشود. همچنین، این روش برای داده‌هایی که تعداد دسته‌ها کم است مناسب‌تر است، زیرا با افزایش تعداد دسته‌ها، تعداد ویژگی‌های جدید نیز افزایش می‌یابد و ممکن است منجر به افزایش بعد داده‌ها شود.

Label Encoding:

در این روش، به هر دسته از داده‌های رشته‌ای یک عدد صحیح نسبت می‌دهیم. این روش مناسب است وقتی ترتیب دسته‌ها اهمیت دارد و می‌خواهیم مدل ماشین‌لرنینگ از این ترتیب استفاده کند. برای مثال، اگر دسته‌ها به صورت "کم"، "متوسط" و "زیاد" باشند، می‌توانیم به آن‌ها به ترتیب اعداد 0، 1 و 2 نسبت دهیم. این روش برای داده‌هایی که تعداد دسته‌ها زیاد است مناسب‌تر است، زیرا تعداد ویژگی‌های جدید را افزایش نمی‌دهد.

اگر ترتیب دسته‌ها اهمیت دارد و تعداد دسته‌ها زیاد است لیبل انکدینگ را انتخاب می‌کنیم.

 اگر ترتیب دسته‌ها اهمیت ندارد و تعداد دسته‌ها کم است وان هات انکدینگ را انتخاب می‌کنیم.

برای فیچر های زیر چون تعداد دسته ها کم و ترتیب اهمیت ندارد از وان هات انکدینگ استفاده میکنیم

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn import decomposition
```

```python
data_new = train_df
pca = PCA(n_components=2)
pca.fit(data_new )
pca_train = pca.transform(data_new )
PC1 = pca_train[:,0]
PC2 = pca_train[:,1]
Income = data_new .Income
label = data_new .Income
```

```python
labels=[]
for points in Income:
  labels.append(label[points])

zipped = list(zip(PC1
                    ,PC2
                    , Income
                    , labels))
pc_df = pd.DataFrame(zipped,
                       columns= ['PC1',
                                 'PC2',
                                 'Income',
                                 'Label'])
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(
      x="PC1",
      y="PC2",
      hue="Income",
      palette=sns.color_palette("hls", 10),
      data=pc_df,
      legend="full",
      alpha=0.3
)
plt.title("Figure 1: Scatter Plot",
          fontsize=12)
plt.xlabel('First Principal Component',
           fontsize=12)
plt.ylabel('Second Principal Component',
           fontsize=12)
```

loss{'hinge', 'squared_hinge'}, default='squared_hinge'
تابع ضرر را مشخص می کند. "hinge" افت استاندارد SVM است (به عنوان مثال توسط کلاس SVC استفاده می شود) در حالی که "squared_hinge" مربع افت لولا است. ترکیب پنالتی='l1' و loss='hige' پشتیبانی نمی شود.

multi_class{'ovr', 'crammer_singer'}, default='ovr'
اگر y بیش از دو کلاس داشته باشد، استراتژی چند کلاسه را تعیین می کند. "ovr" n_classes طبقه بندی کننده های یک در مقابل ریلکس را آموزش می دهد، در حالی که "crammer_singer" یک هدف مشترک را در تمام کلاس ها بهینه می کند.

tolfloat، پیش فرض = 1e-4
معیارهای توقف.

tolfloat, default=1e-3
Tolerance for stopping criterion.



loss{‘hinge’, ‘squared_hinge’}, default=’squared_hinge’
Specifies the loss function. ‘hinge’ is the standard SVM loss (used e.g. by the SVC class) while ‘squared_hinge’ is the square of the hinge loss.

هرچند ما اینجا مقدار پارامترها را مورد بررسی قرار دادیم طبق تعریف  تولرانس یا مقدار خطا واضح بود که ده به توان منفی چهار مقدار دقیق تری را خروجی میدهد

```python
#hame rahehal ha ra hesab mikonad
#penalty = ['l1','l2']
loss_1 = ['hinge', 'squared_hinge']
loss = 'hinge'
multi_class = ['ovr', 'crammer_singer']
tol = [1e-2, 1e-3]
```
